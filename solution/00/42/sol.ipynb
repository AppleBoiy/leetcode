```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt

_trap = [0, 1, 0, 2, 1, 0, 1, 3, 2, 1, 2, 1]


def display_trap(trap):
    trap_len = len(trap)
    data = pd.DataFrame({'height': trap, 'index': range(trap_len)})
    sns.set(style="whitegrid")
    fig, ax = plt.subplots(figsize=(trap_len, max(trap)))  # Adjust figure size dynamically based on data
    sns.barplot(x='index', y='height', data=data, ax=ax, width=1.0)  # Set bar width to 1.0
    plt.yticks(range(0, max(trap) + 1))
    ax.set_xlim(-0.5, trap_len - 0.5)
    ax.set_ylim(0, max(trap) + 1)
    ax.set_aspect(aspect='equal')
    plt.show()
```

```python
_trap = [0, 1, 0, 2, 1, 0, 1, 3, 2, 1, 2, 1]
display_trap(_trap)
```

```python
def trap_rain(trap):
    trap_len = len(trap)
    left_max = [0] * trap_len
    right_max = [0] * trap_len
    left_max[0] = trap[0]
    right_max[-1] = trap[-1]
    for i in range(1, trap_len):
        left_max[i] = max(left_max[i - 1], trap[i])
    for i in range(trap_len - 2, -1, -1):
        right_max[i] = max(right_max[i + 1], trap[i])
    trapped_water = 0
    filled_trap = [min(left_max[i], right_max[i]) for i in range(trap_len)]
    for i in range(trap_len):
        trapped_water += filled_trap[i] - trap[i]
    return trapped_water, left_max, right_max, filled_trap
```

```python
trapped_water, left_max, right_max, filled_trap = trap_rain(_trap)
print(f'Trapped water: {trapped_water}')
```
